```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
import plotly.express as px
```

```python
# Import the data
data = pd.read_csv("superstore (1).csv")

# Create a SQLite database in memory
conn = sqlite3.connect(":memory:")

# Load DataFrame into a SQL table
data.to_sql("table", conn, index=False, if_exists="replace")

# Display the Dataframe with SQL query
query = 'SELECT * FROM "table"'
df = pd.read_sql_query(query, conn)
print(df)
```

```python
# Print the number of clients
query = 'SELECT DISTINCT "Customer Name" FROM "table"'
df = pd.read_sql_query(query, conn)
print(df)
```

```python
# Create a Table with number of order for each day
query = 'SELECT "Order Date", COUNT("Order ID") as "Number of Orders" FROM "table" GROUP BY "Order Date"'
df = pd.read_sql_query(query, conn)
print(df)
```

```python
# Print the distribution of the number of orders
df["Order Date"] = pd.to_datetime(df["Order Date"], format="mixed")
sns.histplot(
    df["Order Date"],
    bins=48,
    kde=True,
    color="blue",
)
plt.title("Distribution of Orders Over Time")
plt.xlabel("Order Date")
plt.xticks(rotation=45)
plt.ylabel("Density")
plt.show()
```

```python
#Profit over GEO
query = 'SELECT "State", SUM("Sales") FROM "table" GROUP BY "State"'
df = pd.read_sql_query(query, conn)
print(df)
```

```python
fig = px.choropleth(df,
                    locations=data["Country"],          # column with state names
                    locationmode='country names',  # tells Plotly these are US states
                    color=data["Sales"],              # column to color by
                    color_continuous_scale='Blues',  # gradient color scale
                    scope='world',                # focus map on USA
                    labels={'sales': 'Sales Amount'}  # label on legend
                   )

fig.show()
```
